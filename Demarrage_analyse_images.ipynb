## üèóÔ∏è Import des packages pour l'analyse d'images


```python
import os
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

## üßÆ Cr√©ation d'une image √† partir d'un tableau de donn√©es

Pour comprendre √† quoi correspond une image, on va cr√©er directement des **images √† partir de tableaux numpy**. Pour la mise en application, on va cr√©er le **[drapeau de l'Ukraine](https://fr.wikipedia.org/wiki/Drapeau_de_l%27Ukraine)** √† partir de z√©ro.

Les couleurs du drapeau sont les suivantes :

- Couleur bleue : #005bbc (R: 0, G: 91, B: 188)
- Couleur jaune : #ffd600 (R: 255, G: 214, B: 0)

```python
# Bleu 0 91 188
top = np.full((75, 225, 3), [0, 91, 188])
plt.imshow(top)
plt.axis("off")
plt.show()
```

```python
# Jaune 255 214 0
bottom = np.full((75, 225, 3), [255, 214, 0])
plt.imshow(bottom)
plt.axis("off")
plt.show()
```

```python
flag = np.concatenate((top, bottom), axis=0)
plt.imshow(flag)
plt.axis("off")
plt.show()
```

## üì∑ D√©composition d'une image

Pour travailler sur une image, on peut utiliser le module [Image](https://pillow.readthedocs.io/en/stable/reference/Image.html) de la biblioth√®que Pillow.

```python
path = os.path.join(os.getcwd(), "content", "chat.png")
image = Image.open(path)
display(image)
```

```python
print(f"Format de l'image : {image.format}\nTaille de l'image : {image.size}")
```

Pour effectuer des **traitements sur l'image**, on peut la convertir en un **tableau Numpy √† 4 dimensions** (3 couleurs primaires et la transparence)

```python
arr = np.array(image)
plt.imshow(arr)
plt.show()
```

```python
arr.shape
```

```python
fig, axs = plt.subplots(1, 4, figsize=(15,15))
for i, color in enumerate(zip(["Reds", "Greens", "Blues", "gray"], ["rouge", "vert", "bleu", "transparence"])):
    axs[i].imshow(arr[:,:, i], cmap=color[0])
    axs[i].set_title(f"Image avec le channel {color[1]}")
    axs[i].axis('off')
plt.show()
```
